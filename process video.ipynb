```python
def process_video(video,crop_loc):
    fps=video.get(5)
    total=video.get(7)
    frames=[]
    index=[]
    step=fps*2
    for i in range(int(total/step)):
        frames.append(get_key_part(get_by_frame(video,int(i*step)),crop_loc))
        index.append(int(i*step))
        print("\r"+str(round(i/int(total/step),3)),end="")
    print()
    return np.array(frames),index

def get_crop_loc(video):
    if video.get(3)==1920:
        return [0.12,0.35,0.08]
    elif video.get(3)==1280:
        return [0.05,0.25,0.1]

def get_key_part(frame,crop_loc):
    h,w=frame.shape[:2]
    return frame[int(crop_loc[0]*h):int(crop_loc[1]*h),int(w*(1/2-crop_loc[2])):int(w*(1/2+crop_loc[2]))] 

def get_by_frame(video,index):
    video.set(1,index)
    _,frame=video.read()
    video.set(1,0)
    return frame

def get_key_by_frame(video,index):
    frame=get_by_frame(video,index)
    crop_loc=get_crop_loc(video)
    key=get_key_part(frame,crop_loc)
    return key

def display_frame(video,index):
    frame=get_by_frame(video,index)
    figure(figsize=(15,12))
    imshow(frame)
    axis("off")
    show()

def display_frame_key(video,index):
    frame=get_by_frame(video,index)
    figure(figsize=(15,12))
    crop_loc=get_crop_loc(video)
    imshow(get_key_part(frame,crop_loc))
    axis("off")
    show()
    
def preprocess(x,y):
    r=[]
    for i in x:
        i=transforms.ToTensor()(i)
        i=transforms.Normalize(mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225])(i)
        r.append(i)
    x=torch.stack(r)
    return x.to("cuda:0"),Tensor(y).long().to("cuda:0")

def resize_key_frame(frame,size=(200,180)):
    return np.array(Image.fromarray(frame).resize(size))

def reshape_batch(batch):
    return np.array([resize_key_frame(i) for i in batch])

def video_to_predictable(path):
    video=cv2.VideoCapture(path)
    crop_loc=get_crop_loc(video)
    data,index=process_video(video,crop_loc)
    return data,index,video



def predict(x):
    batch_size=32
    model.eval()
    with torch.no_grad():
        prediction=[]
        for i in range(0,len(x),batch_size):
            
            tx,ty=preprocess(x[i:i+batch_size],[])
            
            pred=model(tx.float())
            prediction+=list(pred.argmax(1).cpu().numpy())
            print(str(i)+"  "+str(round(i/len(x),3)),end="\r")
    torch.cuda.empty_cache()
    return prediction  


def safe_predict(x,index,video):
    pred=predict(x)
    print(len(pred),np.sum(pred))
    for i in range(len(pred)):
        if pred[i]==1 and test_neighbor(video,index[i])==0:
            pred[i]=0
        print(round(i/len(pred),3),end="\r")
    return pred
    
    
def test_neighbor(video,k,radius=20):
    crop_loc=get_crop_loc(video)
    start=k-radius
    video.set(1,start)
    x=[]
    for i in range(2*radius):
        _,frame=video.read()
        x.append(get_key_part(frame,crop_loc))
    
    x=np.array(x)
    x=normalize(x)
    pred=predict(x)
    l=np.sum(pred[:radius])
    r=np.sum(pred[radius:])
    if l>r:
        if l>radius/2:
            return 1
        else:
            return 0
    else:
        if r>radius/2:
            return 1
        else:
            return 0

 
```
